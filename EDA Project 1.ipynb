## **Univariate Analysis on Numerical and Categorical Columns**

### Univariate Non-visual Analysis on Numerical Columns

```python
# Calculating min, max, std, mean, median, IQR for each column
numerical_columns = df.select_dtypes(include=['int64', 'float64'])
numerical_description = numerical_columns.describe()
print(numerical_description)
```

```python
for col in numerical_columns:
    print(f"\nAnalysis for column: {col}")
    print(f"Mean: {numerical_columns[col].mean()}")
    print(f"Median: {numerical_columns[col].median()}")
    print(f"Standard Deviation: {numerical_columns[col].std()}")
    print(f"Variance: {numerical_columns[col].var()}")
    print(f"Min: {numerical_columns[col].min()}")
    print(f"Max: {numerical_columns[col].max()}")
```

```python
# Skewness and Kurtosis
# Skewness and Kurtosis for numerical columns
for col in numerical_columns:
    skewness = numerical_columns[col].skew()
    kurtosis = numerical_columns[col].kurtosis()
    print(f"\nSkewness of {col}: {skewness}")
    print(f"Kurtosis of {col}: {kurtosis}")
```

## **Research Question**

Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you.
Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)


```python
df.columns
```

```python
cs_grads = df[df['specialization'] == 'Computer Science Engineering']

job_roles = ['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer']
cs_grads_filtered = cs_grads[cs_grads['designation'].isin(job_roles)]

average_salary = cs_grads_filtered['salary'].mean()
print(f"Average Salary: {average_salary} Lakhs")

lower_limit = 2.5
upper_limit = 3.0

if lower_limit <= average_salary <= upper_limit:
    print("The average salary falls within the claimed range of 2.5-3 Lakhs.")
else:
    print("The average salary does NOT fall within the claimed range of 2.5-3 Lakhs.")
```

```python
from scipy.stats import ttest_1samp

test_result = ttest_1samp(cs_grads_filtered['salary'], popmean=2.75) 

print(test_result)

if test_result.pvalue < 0.05:
    print("We reject the null hypothesis. The salary is significantly different from 2.75 Lakhs.")
else:
    print("We fail to reject the null hypothesis. The salary is not significantly different from 2.75 Lakhs.")
```

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['gender'], df['specialization'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2}")
print(f"P-value: {p}")

if p < 0.05:
    print("There is a significant relationship between gender and specialization (dependent).")
else:
    print("There is no significant relationship between gender and specialization (independent).")
```
